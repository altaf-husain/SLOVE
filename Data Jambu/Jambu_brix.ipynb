```python
from sklearn.ensemble import RandomForestRegressor
from micromlgen import port
import pandas as pd

# Membaca file CSV
df = pd.read_csv('Jambu_brix.csv', header=0)

names = df.columns.tolist()

# Memisahkan kolom menjadi target dan fitur
y = df.iloc[:, 0]  # Kolom 1 sebagai target
X = df.iloc[:,0:]  # Kolom 2-21 sebagai fitur

clfb = RandomForestRegressor(n_estimators= 60)
clfb.fit(X, y)
print(port(clfb))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Function to calculate RPD
def calculate_rpd(y_true, y_pred):
    std_dev = y_true.std()
    rpd = std_dev / mean_squared_error(y_true, y_pred, squared=False)
    return rpd

# Random Forest Regression
pred = clf.predict(X)
random_forest_r2 = r2_score(y, pred)
print(f"Random Forest R2: {random_forest_r2}")
random_forest_rpd = calculate_rpd(y, pred)
print(f"Random Forest RPD: {random_forest_rpd}")
```
